## Key Findings
1. **XGBoost** best overall: Acc=88.62%, F1=0.7559, AUC=0.8867
2. **LR** safest: highest recall (72.81%), fewest FN (155, FN-rate=27.2%)
3. **Most accurate != safest** -- accuracy alone is insufficient for clinical trust
4. **XGBoost best calibrated** (Brier=0.0960); LR worst (Brier=0.1352)
5. **SHAP**: RR interval, QRS count, ventricular rate are top clinically meaningful drivers
